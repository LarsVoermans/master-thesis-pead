```python
import pandas as pd

# CSV with earnings & returns
eps_df = pd.read_csv("EPSandValueChange.csv")

# Excel with financial features
fin_df = pd.read_excel(r"C:\Data science and society\Afstuderen\Financial_Data.xlsx")
```

```python
fin_df = fin_df.rename(columns={'Ticker Symbol': 'Ticker'})
```

```python
# Ticker
eps_df['Ticker'] = eps_df['Ticker'].str.strip().str.upper()
fin_df['Ticker'] = fin_df['Ticker'].str.strip().str.upper()

# Date
eps_df['EarningsDate'] = pd.to_datetime(eps_df['EarningsDate'])
fin_df['Date'] = pd.to_datetime(fin_df['Date'])  # pas aan als kolom anders heet
```

```python
eps_df['Year'] = eps_df['EarningsDate'].dt.year
eps_df['Month'] = eps_df['EarningsDate'].dt.month

fin_df['Year'] = fin_df['Date'].dt.year
fin_df['Month'] = fin_df['Date'].dt.month
```

```python
merged = pd.merge(
    eps_df,
    fin_df,
    on=['Ticker', 'Year', 'Month'],
    how='inner'
)
```

```python
merged.to_csv("Offical_Dataset.csv", index=False)
```
